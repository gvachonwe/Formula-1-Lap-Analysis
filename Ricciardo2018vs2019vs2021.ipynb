```python
import fastf1 as ff1
from fastf1 import plotting
import matplotlib.pyplot as plt

# Plot setup

plotting.setup_mpl()
ff1.Cache.enable_cache('cache')

# Load sessions

session2018 = ff1.get_session(2018, 'Monaco GP', 'Q')
session2018.load()
session2019 = ff1.get_session(2019, 'Monaco GP', 'Q')
session2019.load()
session2021 = ff1.get_session(2021, 'Monaco GP', 'Q')
session2021.load()

lap2018 = session2018.laps.pick_driver('RIC').pick_fastest()
tel2018 = lap2018.get_car_data().add_distance()

lap2019 = session2019.laps.pick_driver('RIC').pick_fastest()
tel2019 = lap2019.get_car_data().add_distance()

lap2021 = session2021.laps.pick_driver('RIC').pick_fastest()
tel2021 = lap2021.get_car_data().add_distance()

# Initializing plot
fig, ax = plt.subplots(figsize=(15,8))

# Format plot

colourRenault = ff1.plotting.team_color('AST')
colourMcLaren = ff1.plotting.team_color('MCL')
colourRedBull = ff1.plotting.team_color('RBR')

ax.plot(tel2018['Distance'], tel2018['Speed'], color = colourRedBull, label = '2018')
ax.plot(tel2019['Distance'], tel2019['Speed'], color = colourRenault, label = '2019')
ax.plot(tel2021['Distance'], tel2021['Speed'], color = colourMcLaren, label = '2021')

ax.set_xlabel('Distance [m]')
ax.set_ylabel('Speed [km/h]')
ax.set_title('Ricciardo Monaco Qualifying Lap Comparison (Distance vs. Speed)')
ax.legend()
plt.show()
```
